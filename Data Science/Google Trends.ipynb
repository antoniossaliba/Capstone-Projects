```python
date_array = df_tesla["MONTH"].to_numpy()
```

```python
price = df_tesla["TSLA_USD_CLOSE"].to_numpy()
trend = df_tesla["TSLA_WEB_SEARCH"].to_numpy()
```

```python
fig, ax1 = plt.subplots(figsize=(14, 8))
ax1.set_title("Tesla Web Search vs Price")
ax1.set_ylabel("TSLA Stock Price", color="red")
ax1.plot(date_array, price, color="red", linewidth=3)
ax1.tick_params(axis='x', labelsize=14)
plt.xticks(rotation=45)

ax2 = ax1.twinx()
ax2.set_ylabel("Search Trend", color="blue")
ax2.plot(date_array, trend, color="blue", linewidth=3)


plt.gcf().set_dpi(100)
plt.show()
```

```python
trend = df_bitcoin_search["BTC_NEWS_SEARCH"].to_numpy()
```

```python
date_array = df_bitcoin_search["MONTH"].to_numpy()
```

```python
fig, ax1 = plt.subplots(figsize=(14, 8))
ax1.set_title("Bitcoins News Search vs Resampled Price")
ax1.set_ylabel("BTC Price", color="red")
ax1.plot(date_array, price, color="red", linewidth=1)
ax1.tick_params(axis='x', labelsize=5)
plt.xticks(rotation=45)

ax2 = ax1.twinx()
ax2.set_ylabel("Search Trend", color="blue")
ax2.plot(date_array, trend, color="blue", linewidth=1)

plt.show()
```

```python
date_array = df_unemployment1["MONTH"].to_numpy()
rate = df_unemployment1["UNRATE"].to_numpy()
search = df_unemployment1["UE_BENEFITS_WEB_SEARCH"].to_numpy()
```

```python
fig, ax1 = plt.subplots(figsize=(14, 8))
ax1.set_title("Monthly Search of 'Unemployment Benefits' in the U.S. vs the U/E Rate")
ax1.set_ylabel("FRED U/E Rate", color="red")
ax1.plot(date_array, rate, color="red", linewidth=3)
ax1.tick_params(axis='x', labelsize=5)
plt.xticks(rotation=45)

ax2 = ax1.twinx()
ax2.set_ylabel("Search Trend", color="blue")
ax2.plot(date_array, search, color="blue", linewidth=3)

plt.show()
```

```python
rolled = df_unemployment1.rolling(6)["UE_BENEFITS_WEB_SEARCH"].mean().to_numpy()
```

```python
rate = df_unemployment1.rolling(6)["UNRATE"].mean().to_numpy()
```

```python
date_array = df_unemployment1["MONTH"].to_numpy()
```

```python
fig, ax1 = plt.subplots(figsize=(14, 8))
ax1.set_title("Rolling Monthly US 'Unemployment Benefits' Web Searches vs UNRATE")
ax1.set_ylabel("FRED U/E Rate", color="red")
ax1.plot(date_array, rolled, color="red", linewidth=3)
ax1.tick_params(axis='x', labelsize=5)
plt.xticks(rotation=45)

ax2 = ax1.twinx()
ax2.set_ylabel("Search Trend", color="blue")
ax2.plot(date_array, rate, color="blue", linewidth=3)

plt.show()
```

```python
df_unemployment2["MONTH"] = pd.to_datetime(df_unemployment2["MONTH"])
date_array = df_unemployment2["MONTH"].to_numpy()
search = df_unemployment2["UE_BENEFITS_WEB_SEARCH"].to_numpy()
rate = df_unemployment2["UNRATE"].to_numpy()
```

```python
fig, ax1 = plt.subplots(figsize=(14, 8))
ax1.set_title("Monthly US 'Unemployment Benefits' Web Search vs UNRATE incl 2020")
ax1.set_ylabel("FRED U/E Rate", color="red")
ax1.plot(date_array, rate, color="red", linewidth=2)
ax1.tick_params(axis='x', labelsize=5)
plt.xticks(rotation=45)

ax2 = ax1.twinx()
ax2.set_ylabel("Search Trend", color="blue")
ax2.plot(date_array, search, color="blue", linewidth=2)

plt.show()
```
