```python
df_free_vs_paid = df_apps_clean.groupby(["Category", "Type"], as_index=False).agg({'App': pd.Series.count})
df_free_vs_paid.head()
```

```python
for_free_counts = df_free_vs_paid[df_free_vs_paid["Type"] == "Free"].groupby("Category").sum()["App"].to_numpy()
for_money_counts = df_free_vs_paid[df_free_vs_paid["Type"] == "Paid"].groupby("Category").sum()["App"].to_numpy()
categories1 = df_free_vs_paid[df_free_vs_paid["Type"] == "Paid"].groupby("Category").sum().index.to_numpy()
categories2 = df_free_vs_paid[df_free_vs_paid["Type"] == "Free"].groupby("Category").sum().index.to_numpy()

plot = po.Figure(data=[po.Bar(name="Type=Free", x=categories2, y=for_free_counts), (po.Bar(name="Type=Paid", x=categories1, y=for_money_counts))])
plot.update_xaxes(categoryorder='total descending')
plot.show()
```

```python
free_installations_count = df_apps_clean[df_apps_clean["Type"] == "Free"]["Installs"].to_numpy()
paid_installations_count = df_apps_clean[df_apps_clean["Type"] == "Paid"]["Installs"].to_numpy()
```

```python
fig = go.Figure()

fig.add_trace(go.Box(y=free_installations_count, name="Free"))
fig.add_trace(go.Box(y=paid_installations_count, name="Paid"))

fig.update_layout(yaxis_range=[0, 100100000])

fig.show()
```

```python
df_paid_infos = df_apps_clean.where(df_apps_clean["Type"] == "Paid").dropna()
categories = df_paid_infos.groupby("Category").count().index.to_numpy()
```

```python
fig = go.Figure()

for category in categories:
  fig.add_trace(go.Box(y=df_paid_infos["Revenue_Estimate"].where(df_paid_infos["Category"] == category).dropna(), name=category))

fig.update_layout(yaxis_range=[0, 10000000])

fig.show()
```
